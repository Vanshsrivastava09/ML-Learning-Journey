```python
# Challenge

import numpy as np

salary = np.array([25000, 30000, 35000, 40000, 45000, 50000, 200000])

print(salary)

print(salary.mean())

print(np.median(salary))

print(salary.max() - salary.min())

print(np.var(salary))

print(np.std(salary))

import matplotlib.pyplot as plt

plt.hist(salary)
plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.show()

print("Mean:", salary.mean())
print("Median:", np.median(salary))

## BOXPLOT

import matplotlib.pyplot as plt

plt.boxplot(salary)
plt.title("Salary Boxplot")
plt.show()

Q1 = np.percentile(salary, 25)
Q3 = np.percentile(salary, 75)

print("Q1:", Q1)
print("Q3:", Q3)

IQR = Q3 - Q1

print("IQR:", IQR)

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)

outliers = salary[(salary < lower_bound) | (salary > upper_bound)]

print("Outliers:", outliers)
```
