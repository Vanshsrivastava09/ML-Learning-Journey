```python
import pandas as pd

data = {
    "Name": ["Vansh", "Rahul", "Aman", "Priya"],
    "Marks": [90, 75, 88, 95],
    "Attendance": [92, 80, 85, 98] 
    }

df = pd.DataFrame(data)

print(df)

print(df.head()) # view first rows
print(df.shape) # (rows, columns)
print(df.columns) # column names
print(df.info()) # dataset info

print(df["Marks"])
print(df[["Name", "Marks"]])


# Filtering
# Students scoring above 85
high_scorers = df[df["Marks"]>85]
print(high_scorers)

# Sorting
print(df.sort_values(by="Marks", ascending=False))

# Basic Analysis
print(df["Marks"].mean())
print(df["Marks"].max())
print(df["Marks"].min())

# Add new Column
df["Bonus"] = 5
print(df)

# Final Insights
print("Observations:")
print("-Priya scored highest")
print("-Average marks are strong")
print("-Attendance appears positively related to marks")


```

```python
# Employee Dataset Analysis

import pandas as pd

employees = {
    "Name": ["Aman", "Riya", "Karan", "Neha"],
    "Salary": [50000, 65000, 48000, 72000],
    "Experience": [2, 5, 1, 6]
}

# 1. Convert to DataFrame

df = pd.DataFrame(employees)
print(df)

# 2. Find average salary

average_salary = df["Salary"].mean()
print(f"Average Salary: {average_salary:.2f}")

# 3. Find employees with highest salary

highest_salary = df["Salary"].max()
print(f"Highest Salary: {highest_salary}")
highest_paid_employees = df[df["Salary"] == highest_salary]
print("Employees with Highest Salary:")
print(highest_paid_employees)

#  4. FInd employees with experience>2
experienced_employees = df[df["Experience"] > 2]
print("\nEmployees with Experience > 2 years:")
print(experienced_employees)

# Sort salaries in descending order
sorted_employees = df.sort_values(by="Salary", ascending=False)
print("\nEmployees sorted by Salary (Descending):")
print(sorted_employees)

# Add bonus column 10000
df["Bonus"] = 10000
print(df)

print("\nInsights:")
print("- Neha has the highest salary.")
print("- Average salary is balanced.")
print("- Experienced employees generally earn higher salaries.")
```

```python
# Student Data Analysis

import pandas as pd

students = {
    "Name": ["Aman", "Priya", "Rahul", "Sneha", "Karan"],
    "Math": [78, 92, 65, 88, 70],
    "Science": [85, 95, 70, 90, 75],
    "English": [80, 89, 60, 92, 72]
}

# Create DataFrame

df = pd.DataFrame(students)
print(df)

# Add new Column: Toatal Marks of all subjects
df["Total"] = df["Math"] + df["Science"] + df["English"]
print(df)

# Add another column: Average Marks per student
df["Average"] = df["Total"]/3
print(df)

# Find Topper
topper = df[df["Average"] == df["Average"].max()]
print("\nTopper:")
print(topper)

# Filter Students with average > 80
high_achievers = df[df["Average"]>80]
print("\nStudents with Average Marks > 80:")
print(high_achievers)

# Sort by Total Marks descending
sorted_students = df.sort_values(by="Total", ascending=False)
print("\nStudents sorted by Total Marks (Descending):")
print(sorted_students)

# Add Grade Column
def assign_grade(avg):
    if avg >= 90:
        return "A"
    elif avg >= 80:
        return "B"
    elif avg >= 70:
        return "C"
    else:
        return "D"
df["Grade"] = df["Average"].apply(assign_grade)
print(df)
```

```python
# CSV files + missing values

import pandas as pd

students_data = {
    "Name": ["Aman", "Priya", "Rahul", "Sneha", "Karan"],
    "Marks": [78, 92, None, 88, 70],
    "Attendance": [85, 95, 75, None, 80]
}

df = pd.DataFrame(students_data)
print(df)

# Check missing values
print(df.isnull())

# Count missing values
print(df.isnull().sum())

# Fill missing values
df["Marks"] = df["Marks"].fillna(df["Marks"].mean())
df["Attendance"] = df["Attendance"].fillna(df["Attendance"].mean())
print("\nAfter filling missing values:")
print(df)

# Save Dataset
df.to_csv("cleaned_students_data.csv", index=False)

# Read CSV File
new_df = pd.read_csv("cleaned_students_data.csv")
print(new_df)
```

```python
# Another Challenge

import pandas as pd

products = {
    "Product": ["Laptop", "Phone", "Tablet", "Monitor"],
    "Price": [70000, None, 30000, 15000],
    "Rating": [4.5, 4.7, None, 4.2]
}

# Tasks: Create DataFrame, detect missing values, count missing values, fill missing prices using average, fill missing ratings using avergae, save CSV, read CSV again

df = pd.DataFrame(products)
print(df)

print("\n Missing Values:")
print(df.isnull())

print("\n Count missing values:")
print(df.isnull().sum())

print("\n Fill missing values using average:")
df["Price"] = df["Price"].fillna(df["Price"].mean())
df["Rating"] = df["Rating"].fillna(df["Rating"].mean())
print(df)

print("\n Saved to CSV")
df.to_csv("Cleaned_products_data.csv", index = False)

print("\nReading CSV again:")
new_df = pd.read_csv("Cleaned_products_data.csv")
print(new_df)
```

```python
# Smartphone Sales Analysis

import pandas as pd

phones = {
    "Brand": ["Apple", "Samsung", "OnePlus", "Xiaomi", "Apple", "Samsung"],
    "Price": [80000, 65000, 50000, 30000, 90000, 70000],
    "Rating": [4.8, 4.5, 4.4, 4.2, 4.9, 4.6],
    "Sales": [120, 150, 90, 200, 100, 170]
}

df = pd.DataFrame(phones)
print(df)

df.info() # show dateset info

# Find average phone price, highest rating, total sales
average_price = df["Price"].mean()
print(f"\nAverage Price:")
print(average_price)

highest_rating = df["Rating"].max()
print(f"\nHighest Rating:")
print(highest_rating)

total_sales = df["Sales"].sum()
print(f'\nTotal sales:')
print(total_sales)

# Find most expensive phones
most_expensive =df[df["Price"] == df["Price"].max()]
print("\n Most Expensive Phones:")
print(most_expensive)

# Filter phones wirh rating > 4.5
highly_rated = df[df["Rating"]>4.5]
print("\n Highly Rated Phones:")
print(highly_rated)

# Filter phones with price above 60000
premium_phones = df[df["Price"]>60000]
print("\n Premium Phones:")
print(premium_phones)

# Sort by sales descending
sorted_phone = df.sort_values(by="Sales", ascending=False)
print("\n Phones sorted by Sales (Descending):")
print(sorted_phone)

# Create new column: Revenue = Price * Sales
df["Revenue"] = df["Price"] * df["Sales"]
print(df)

# Find brand with highest total sales
sales_by_brand = df.groupby("Brand")["Sales"].sum()
print("\n Total Sales by Brand:")
print(sales_by_brand)

# Visualize Sales 

import matplotlib.pyplot as plt

plt.bar(df["Brand"], df["Sales"], color="blue")
plt.xlabel("Brand")
plt.ylabel("Sales")
plt.title("Smartphone Sales by Brand")
plt.show()

# Correlation Analysis
print(df.corr(numeric_only=True))

# Heatmap Visualization

import seaborn as sns

sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```
