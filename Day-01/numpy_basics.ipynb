```python
import numpy as np 

arr=np.array([1,2,3,4,5])
print(arr)

print(arr[0]) #first element
print(arr[1]) #second element

print(arr + 10) #add 10 to each element
print(arr * 2) #multiply each element by 2

#statistics
print(arr.mean())
print(arr.max())
print(arr.min())
print(arr.sum())

#reshaping
print(arr.reshape(5,1)) #reshape to 5 rows and 1 column
```

```python
# Matplotlib Basics

import matplotlib.pyplot as plt

sales = np.array([[200, 220, 250], [150, 180, 210], [300, 310, 330]])

total_sales = sales.sum(axis=1)
print(total_sales)

products = ["Product A", "Product B", "Product C"]
plt.bar(products, total_sales)
plt.title("Total sales per Product")
plt.xlabel("Products")
plt.ylabel("Sales")
plt.show()

months = ["Jan", "Feb", "Mar"]
plt.plot(months, sales[0])
plt.title("Sales of Product A over 3 months")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.show()
```

```python
import numpy as np

# Generate random data

random_marks = np.random.randint(40, 100, size=20)

print(random_marks)

# Analyze random data

print("Average:", random_marks.mean())
print("Highest:", random_marks.max())
print("Lowest:", random_marks.min())

# Visualize distribution

import matplotlib.pyplot as plt

plt.hist(random_marks)
plt.title("Distribution of Student marks")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()

# Observation:
print("Most students scored between 60 and 90")
print("Very few low scores")
print("Distribution looks balanced")

# Shape

print(random_marks.shape)
```

```python
# Challenge: 50 random attendance values between 60 and 100

import numpy as np

attendance = np.random.randint(60, 100, size=50)
print(attendance)

print("Average:", attendance.mean())
print("Highest:", attendance.max())

# Plot Histogram

import matplotlib.pyplot as plt

plt.hist(attendance)
plt.title("Distribution of Student Attendance")
plt.xlabel("Attendance")
plt.ylabel("Frequency")
plt.show()
```
